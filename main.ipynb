<h1>1.1 - Scatter Plots


#####
 Individual data point

```python
import numpy as np
import matplotlib.pyplot as plt

years = [2006 + x for x in range(16)]

weights = [80, 83, 84, 85, 86, 87 , 88, 79, 86, 96, 78, 89, 77, 77, 98, 87]

plt.scatter(years, weights, c='red', s=150, marker='*', alpha=0.4
            )
```

# 1.6 - Pie

```python
langs =["c++", "C#", "Python", "Java", "Go"]
votes = [50, 24, 14, 6, 17]
explodes = [0.2, 0, 0, 0, 0.2]
plt.pie(votes, labels=langs, explode=explodes, autopct="%.2f%%", pctdistance=1.5)
```

# 2.1 Plot customization

```python
years = [2014 + x for x in range(8)]

income = [55, 56, 62, 61, 72, 72, 73, 75]

income_ticks = list(range(50, 81, 2))

plt.plot(years, income)
plt.title("Years and dolar")
plt.xlabel("Years")
plt.ylabel("Milions of dollar")

plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks])
```

# 2.2 Legends and multi plots

```python
stock_a = [100,20, 32, 23, 34, 64, 45, 53, 34]
stock_b = [23 ,43, 87, 76, 56, 34, 43, 65, 54]
stock_c = [32, 23, 65, 75, 98, 65, 34, 23, 53]

plt.plot(stock_a, label="Stock a")
plt.plot(stock_b, label="Stock b")
plt.plot(stock_c, label="Stock c")

plt.legend(loc="upper right")
```
